```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [200,300,500,800,1000]
y = [100,200,350,500,600]
```

```python
coeff = np.polyfit(x,y,1)

A = coeff[0]
B = coeff[1]
```

```python
x2 = np.linspace(0,1000,20)
y2 = A*x2 + B
```

```python
plt.plot(x2,y2,"r")
plt.scatter(x,y)
plt.grid()
```

2 degree

```python
np.polyfit(x,y,2)
```

```python
coeff2 = np.polyfit(x,y,2)
a = coeff2[0]
b = coeff2[1]
c = coeff2[2]
```

```python
x3 = np.linspace(0,1000,20)
y3 = a*x3**2 + b*x3 + c
```

```python
plt.plot(x3,y3,"g")
plt.scatter(x,y)
plt.grid()
```

Quiz: find price in year 8

```python
year = [0,1,2,3,4,5]
price = [51000,47000,45000,42500,41000,38000]
```

```python
coeff_q = np.polyfit(year,price,1)

AQ = coeff_q[0]
BQ = coeff_q[1]

year_predict = np.linspace(0,10,20)
price_predict = AQ*year_predict + BQ
```

```python
print(AQ*8 + BQ)
```

```python
plt.scatter(year,price)
plt.plot(year_predict,price_predict,"r")
plt.grid()

x_mark = 8
y_mark = AQ*8 + BQ
plt.scatter(x_mark, y_mark, color='red', label=f"Point ({x_mark}, {y_mark})")
```

```python
coeff_q = np.polyfit(year,price,2)

AQ2 = coeff_q[0]
BQ2 = coeff_q[1]
CQ2 = coeff_q[2]

year_predict2 = np.linspace(0,10,20)
price_predict2 = AQ2*year_predict2**2 + BQ2*year_predict2 + CQ2

plt.scatter(year,price)
plt.plot(year_predict2,price_predict2,"r")
plt.grid()
```
