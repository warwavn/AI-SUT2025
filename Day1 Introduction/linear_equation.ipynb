```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,20) #Gen 20 value between -10 to 10 
print(x)
```

```python
m = 1
c = 1
```

```python
print(y)
```

```python
plt.plot(x,y)

plt.xlabel('x value')
plt.ylabel('y value')
# plt.title('y = mx + c')
plt.grid()
plt.show()
```
