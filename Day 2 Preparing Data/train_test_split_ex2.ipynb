```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(42) # For reproducibility
data = {
'Feature1': np.random.randn(100),
'Feature2': np.random.randint(1, 10, 100),
'Feature3': np.random.choice(['A', 'B', 'C'], 100),
'Target': np.random.randint(0, 2, 100) # Binary target
}
```

```python
df = pd.DataFrame(data)
df
```

encoder.fit(df['Feature3'])\
encoder.transform(df['Feature3'])

```python
encoder = LabelEncoder()
df['Feature3'] = encoder.fit_transform(df['Feature3'])
```

```python
df
```

```python
x = df[['Feature1','Feature2','Feature3']]
y = df['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
display(X_test)
display(X_train)

display(y_test)
display(y_train)
```

Data Scaling (Standardize Features)

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
display(X_train_scaled)
display(X_test_scaled)
```
