```python
cols_with_nan = df.columns[df.isna().any()].tolist()

numeric_cols = df[cols_with_nan].select_dtypes(include=['number']).columns

# Define a function to determine the best fill value for each column
def best_fill_value(series):
    mode = series.mode()
    mean = series.mean()
    median = series.median()

    if len(mode) == 1 and abs(mode[0] - median) <= abs(mode[0] - mean):
        return median
    else:
        return mean

# Apply the best_fill_value function to each numeric column
for col in numeric_cols:
    df[col] = df[col].fillna(best_fill_value(df[col]))
```

```python
string_cols = df[cols_with_nan].select_dtypes(include=['object']).columns

df[string_cols] = df[string_cols].fillna(method='ffill')
```
