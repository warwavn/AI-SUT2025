```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_excel(r"C:\Users\User\Downloads\test_label_encoder.xlsx")
df
```

```python
cols_to_encode = df.columns.tolist()
print(cols_to_encode)
```

```python
encoders = {}
for col in cols_to_encode:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    encoders[col] = le 

df
```

```python
df.to_csv("labelencoder.csv")
```

```python
df_decoded = df.copy()

# Apply inverse transform to each encoded column
for col in cols_to_encode:
    df_decoded[col] = encoders[col].inverse_transform(df_decoded[col])

print(df_decoded) 
```
