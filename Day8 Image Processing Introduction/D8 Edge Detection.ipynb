```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

Sobel

```python
img = cv2.imread("kid.jpg",0)

sobelX = cv2.Sobel(img,-1,1,0)                  # (img file, -1: from original img, x-axis, y-axis)
sobelY = cv2.Sobel(img,-1,0,1)
sobelXY = cv2.bitwise_or(sobelX,sobelY)         #or

cv2.imshow("Original",img)
cv2.imshow("X", sobelX)
cv2.imshow("Y", sobelY)
cv2.imshow("XY", sobelXY)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
titles = ["Original","sobelX", "sobelY", "sobelXY"]
images = [img, sobelX, sobelY, sobelXY]

# Plot the images
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

for i, ax in enumerate(axes.flat):
    ax.imshow(images[i], cmap='gray')  # Display image in grayscale
    ax.set_title(titles[i])
    ax.set_xticks([])  # Hide x-axis ticks
    ax.set_yticks([])  # Hide y-axis ticks

plt.tight_layout()  # Ensure the layout doesn't overlap
plt.show()
```

Laplacian

```python
lap = cv2.Laplacian(img,-1)

cv2.imshow("Laplacian",lap)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

Canny

```python
can = cv2.Canny(img,20,200)                 # (img, th, th)

cv2.imshow("Canny",can)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
