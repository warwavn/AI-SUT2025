Multi-variable Linear Regression\
x1, x2, x3,...,xn and y ; one target

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib as plt
from sklearn.metrics import mean_squared_error, r2_score
```

```python
d = [[8, 4, 45],
     [7, 5, 44],
     [8, 6, 50],
     [6, 6, 43],
     [9, 5, 45],
     [8, 3, 44],
     [9, 4, 40],
     [6, 5, 43]]
```

```python
d = np.array(d)
```

```python
x = d[:,0:2]    #col 0-1
y = d[:,2]       #col 2
```

```python
model = LinearRegression()

model.fit(x,y)
```

```python
print(f" y = ({model.coef_[0]:,.2f})X1 + ({model.coef_[1]:,.2f})X2 + {model.intercept_:,.2f}")
```

```python
x_predict = [[6,6],[7,6]]
y_predict = model.predict(x_predict)
p_6_6 = '{:.2f}'.format(y_predict[0])
p_7_6 = '{:.2f}'.format(y_predict[1])

print(f"x1 = 6, x2 = 6, y = {p_6_6}")
print(f"x1 = 7, x2 = 6, y = {p_7_6}")

print(f"x1 = 6, x2 = 6, y = {y_predict[0]:,.2f}")
```

```python
score = model.score(x,y)                # x,y

print("R-2 = ",score)
```

```python
y_pred = model.predict(x)
```

```python
mse = mean_squared_error(y,y_pred)
rmse = np.sqrt(mse)

print(f"MSE = {mse:.2f}")
print(f"rmse = {rmse:.2f}")     
```
