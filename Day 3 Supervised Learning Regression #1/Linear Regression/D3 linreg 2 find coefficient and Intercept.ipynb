```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
x = list(range(5,61,5))                                         #feature
y = [8, 42, 102, 113, 148, 206, 202, 230, 360, 354, 427, 432]   #target ; actual
x = np.array(x).reshape(-1,1)

model = LinearRegression()
model.fit(x,y)
```

Prediction

```python
x_predict = [[8],[18],[33],[59],[70],[75],[100]]        #New data
y_predict = model.predict(x_predict)                    #Make prediction
```

```python
plt.scatter(x_predict,y_predict)
plt.grid()
```

```python
for (i, p) in enumerate(x_predict):
    y_pred = '{:.2f}'.format(y_predict[i])
    print(f'x = {p[0]} predicted y = {y_pred}')
```

```python
ic = '{:.2f}'.format(model.intercept_)
ce = '{:.2f}'.format(model.coef_[0])
print(f'Linear equation is: y = ({ce})x {ic}')
print(f'Linear equation is: y = ({model.coef_[0]:.2f}x) {model.intercept_:.2f}')
```

```python
y_pred_all = model.predict(x)
plt.scatter(x,y,label = "actual")                 #actual data
plt.plot(x,y_pred_all,"r",label = "predict")      #predicted data
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```
