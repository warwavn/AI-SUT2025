```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\User\Downloads\kmeans_bmx - kmeans_bmx.csv")
df = df[['bmxleg', 'bmxwaist']].dropna()
with pd.option_context('display.max_rows', 10): display(df)
model = KMeans(n_clusters=5)
model.fit(df)
centroids = model.cluster_centers_
print('Centroids:\n', centroids)
```

```python
cluster = model.predict(df)
cluster_colors = ['purple','green','blue','orange','brown']
data_colors = []

for i in cluster:
    data_colors.append(cluster_colors[i])

plt.figure(figsize=(10, 6))
# plot data points
plt.scatter(df['bmxleg'], df['bmxwaist'], c=data_colors)
# plot centroid points for each cluster
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*',color='red', s=150)    #s = marker size
plt.xlabel("Upper leg (cm)")
plt.ylabel("Waist size (cm)")
plt.show()    
```
