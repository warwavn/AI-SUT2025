EX1

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
x = [[8, 6], [3, 5], [4, 9], [5, 8], [9, 9]] #[[x1, x2], ...]
y = ['yes', 'no', 'no', 'yes', 'yes']
model.fit(x, y)
# Predict when x1=4, x2=4, y=? และ x1=5, x2=5, y=?
x_predict = [[4, 4], [5, 5],[1,8]]
y_predict = model.predict(x_predict)
print('if x = [4, 4] then y =', y_predict[0])
print('if x = [5, 5] then y =', y_predict[1])
print('if x = [1, 8] then y =', y_predict[2])
print()
print('Probability [1-P    P] : ')
print("Probability ['no'  'yes'] : ")

prob = model.predict_proba(x_predict)
print(prob)
```

EX2 - Study hours

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv(r"C:\Users\User\Downloads\logis_hours_study - logis_hours_study.csv")      #0-fail 1-pass
with pd.option_context('display.max_rows',6):display:df
```

```python
df
```

```python
x = np.array(df['hours_study']).reshape(-1,1)
y = df['pass']

model = LogisticRegression()

model.fit(x,y)
```

```python
x_pred = [2.2,3.3,4.4]
x_pred = np.array(x_pred).reshape(-1,1)

y_pred = model.predict(x_pred)

print("Prediction : ")
for i, xp in enumerate(x_pred):
    r = "Fail" if y_pred[i] == 0 else "Pass"
    print(f' study hours: {xp[0]} hrs => {r}')

print()
print('Logistic Function : ') 
ic  = '{:.2f}'.format(model.intercept_[0])
ce = '{:.2f}'.format(model.coef_[0,0])
print(f' P(x) = 1/(1+exp({ic}+{ce}x))')
```

EX3 - Classify Machine Spec

```python
data = pd.read_excel(r"C:\Users\User\Downloads\machine-spec.xlsx")
data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x = data.drop('machine_meets_spec',axis=1)
y = data['machine_meets_spec']

scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
model = LogisticRegression()
model.fit(x, y)
x_predict = [[30, 6], [40, 8], [50, 5], [60, 3]]
x_predict_sc = scaler.transform(x_predict)
y_predict = model.predict(x_predict_sc)
```

```python
print('Prediction:')
for (i, xp) in enumerate(x_predict):
    mp = 'NOK' if y_predict[i] == 0 else 'OK'
    t = f'Machine Age: {xp[0]} Month(s), '
    t += f'Operate: {xp[1]} Hour(s)/Day '
    t += f'=> Meet Spec: {mp}'
    print(t)

print()
print('Logistic Function:')
ic = '{:.2f}'.format(model.intercept_[0])
ce1 = '{:.2f}'.format(model.coef_[0, 0])
ce2 = '{:.2f}'.format(model.coef_[0, 1])
print(f'P(x) = 1 / (1 + exp({ic} + ({ce1})age + ({ce2})hour))')
```
