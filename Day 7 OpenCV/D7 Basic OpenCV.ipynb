5. Access webcam

```python
cap = cv2.VideoCapture(0)

while(True):
    check,frame = cap.read()                        #several capture present with while loop
    cv2.imshow("My Cap",frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):           #close frame follow key
        break             
cap.release()
cv2.destroyAllWindows()    
```

```python
cap = cv2.VideoCapture(0)

while(True):
    check,frame = cap.read()                        #several capture present with while loop
    flip_frame = cv2.flip(frame,1)                  #0 = x, 1 = y
    cv2.imshow("My Cap",flip_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):           #close frame follow key
        break                
cap.release()
cv2.destroyAllWindows()    
```

Import VDO

```python
capv = cv2.VideoCapture(r"shibuya.mp4")

while (capv.isOpened()):
    check,frame = capv.read()
    if check == True:
        cv2.imshow("VDO",frame)
        if cv2.waitKey(1) & 0xFF == ord('x'):           #close frame follow key
            break       
    else:
        break      
capv.release()
cv2.destroyAllWindows()    
```

Resize

```python
import cv2
capv = cv2.VideoCapture(r"shibuya.mp4")

while (capv.isOpened()):
    check,frame = capv.read()
    
    if check == True:
        rz_frame = cv2.resize(frame,(400,400))
        cv2.imshow("VDO",rz_frame)
        if cv2.waitKey(1) & 0xFF == ord('x'):           #close frame follow key
            break       
    else:
        break      
capv.release()
cv2.destroyAllWindows()    
```

Color convert

```python
import cv2
capv = cv2.VideoCapture(r"shibuya.mp4")

while (capv.isOpened()):
    check,frame = capv.read()
    
    if check == True:
        rz_frame = cv2.resize(frame,(400,400))
        # gray_frame = cv2.cvtColor(rz_frame,cv2.COLOR_BGR2GRAY)
        # gray_frame = cv2.cvtColor(rz_frame,cv2.COLOR_BGR2HSV)
        gray_frame = cv2.cvtColor(rz_frame,cv2.COLOR_BGR2RGB)
        cv2.imshow("VDO",gray_frame)
        if cv2.waitKey(1) & 0xFF == ord('x'):           #close frame follow key
            break       
    else:
        break      
capv.release()
cv2.destroyAllWindows()    
```

VDO recorder

```python
cap = cv2.VideoCapture(0)

# cv2.VideoWriter('file name', format, frame rate, size (w, h))
fourcc = cv2.VideoWriter_fourcc(*'MJPG')                            #.MP4
rec = cv2.VideoWriter('My_VDO.mp4',fourcc,100.0,(640,480))          #frame rate => higher is faster

while (cap.isOpened()):
    check, frame = cap.read()

    
    if check == True:
        cv2.imshow("Recording VDO", frame)
        rec.write(frame)
        if cv2.waitKey(1) & 0xFF == ord('a'):
            break
rec.release()
cap.release()
cv2.destroyAllWindows()
```
