```python
x = np.linspace(-10,10,20) #Gen 20 value between -10 to 10 
print(x)
```

```python
# Given points
x1, y1 = -6, -3
x2, y2 = 8, 7

# Calculate the slope (m)
m = (y2 - y1) / (x2 - x1)

# Calculate the y-intercept (c)
c = y1 - m * x1

# Display the results
print(f"The slope (m) is: {m}")
print(f"The y-intercept (c) is: {c}")
```

```python

y = c + m*x
```
